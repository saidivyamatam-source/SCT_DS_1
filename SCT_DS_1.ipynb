```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
uploaded = files.upload()
file = list(uploaded.keys())[0]
```

```python
df = pd.read_excel(file, sheet_name="Data", skiprows=3)
df.head()
```

```python
india = df[df["Country Name"] == "India"]
years = [str(y) for y in range(2000, 2023)]  # ensure year columns as string
population = india[years].values.flatten()
```

```python
plt.figure(figsize=(12,6))
plt.bar(years, population, color="skyblue")
plt.xticks(rotation=45)
plt.title("India's Population Growth (2000–2022)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
if "2020" in df.columns:
    plt.figure(figsize=(10,6))
    sns.histplot(df["2020"].dropna(), bins=30, kde=True, color="orange")
    plt.title("Population Distribution Across Countries (2020)")
    plt.xlabel("Population")
    plt.ylabel("Number of Countries")
    plt.show()
```

```python
if "2020" in df.columns:
    top10 = df[["Country Name", "2020"]].dropna().sort_values(by="2020", ascending=False).head(10)
    plt.figure(figsize=(12,6))
    sns.barplot(x="Country Name", y="2020", data=top10, palette="viridis")
    plt.xticks(rotation=45)
    plt.title("Top 10 Most Populated Countries (2020)")
    plt.xlabel("Country")
    plt.ylabel("Population")
    plt.show()
```
